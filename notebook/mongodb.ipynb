```python
import pymongo
from pymongo.errors import ConnectionFailure, AutoReconnect,BulkWriteError

client = pymongo.MongoClient(CONNECTION_URL, serverSelectionTimeoutMS=5000)
database = client[DB_NAME]
collection = database[COLLECTION_NAME]

batchsize=10000

for i in range(0,len(data_dict),batchsize):

    batch=data_dict[i:i+batchsize]

    try:

        result=collection.insert_many(batch,ordered=False)

    except BulkWriteError as bwe:

        print(f"errors insterting batch {i//batchsize+1}:{bwe.details}")


```
